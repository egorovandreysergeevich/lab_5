5.2. Используя данные об автомобилях, создайте диаграмму рассеяния (*scatter-plot*), на которой *размер* (*size*) точек становится больше при наведении на них курсора.

```python
# Создадим перменную selection, которая выделяет элемент при наведении на него указателя мыши.
selection = alt.selection_single(on='mouseover', nearest=True)
# Создадим перменную chart, которая построит диаграмму рассеяния по набору данных cars с указанием лошадиных сил на оси x и миль на галлон на оси y
chart = alt.Chart(cars).mark_circle().encode(
    x = 'Horsepower',
    y = 'Miles_per_Gallon',
    opacity = alt.value(0.5),
    tooltip = ['Horsepower', 'Miles_per_Gallon'],
# Цвет обозначает регион происхождение автомобиля
    color = 'Origin',
# Размер точки меняется в зависимости от 50 до 500 в зависимости от того, выбрана она или нет
    size = alt.condition(
        selection,
        alt.value(500),
        alt.value(50)
    )
).add_selection(
    selection
# Сделаем диагрумму интерактивной
).interactive()
# Выведим диаграмму
chart
```

5.3. Используя данные об автомобилях, создайте двухпанельную (*two-panel*) гистограмму (количество миль на галлон на одной панели, количество лошадиных сил на другой), где вы можете перетащить мышь, чтобы выбрать данные на левой панели, чтобы отфильтровать данные на второй панели.

```python
# Переменная auto выделяет диапазон значений на обоих графиках одновременно.
auto = alt.selection_interval(encodings=['x'])

# Создадим гистограмму для переменной Miles_per_Gallon
his_m = alt.Chart(cars).mark_bar().encode(
    x=alt.X('Miles_per_Gallon:Q', bin=True),
    y='count()',
    color=alt.condition(auto, alt.value('blue'), alt.value('green'))
).add_selection(
    auto
)

# Создадим гистограмму для переменной Horsepower
his_h = alt.Chart(cars).mark_bar().encode(
    x=alt.X('Horsepower:Q', bin=True),
    y='count()',
    color=alt.condition(auto, alt.value('red'), alt.value('orange'))
).transform_filter(
    auto
)

# Объединим гистограммы в один график
his_m | his_h
```

5.4. Измените приведенный выше пример диаграммы разброса и гистограммы, чтобы

- панорамировать и увеличивать диаграмму рассеяния;
- гистограмма отражала только те точки, которые видны в данный момент.

```python
# Создадим диаграмму рассеяния для переменной Miles_per_Gallon
miles_point = alt.Chart(cars).mark_point().encode(
    x='Miles_per_Gallon',
    y='count()',
    color=alt.condition(auto, alt.value('blue'), alt.value('green'))
).add_selection(
    auto
).interactive()

# Создадим гистограмму для переменной Horsepower
hoursepower_bar = alt.Chart(cars).mark_bar().encode(
    alt.X('Horsepower', bin=True),
    alt.Y('count()'),
    color=alt.condition(auto, alt.value('red'), alt.value('orange'))
).transform_filter(
    auto
).interactive()

# Объединим в один график
miles_point | hoursepower_bar
```

5.5. Создайте диаграмму на основе самостоятельно сгенерированных данных и постройте кривые синуса и косинуса с помощью `transform_calculate`.

x = pd.DataFrame({'x': np.linspace(-len(Имя), len(Отчество))})

Импортируем библиотеки

```python
import pandas as pd
import numpy as np
```

Создаем DataFrame с значениями x

```python
x = pd.DataFrame({'x': np.linspace(-len("6"), len("6"))})
x.head()
```

Рассчитываем синусы и косинусы

```python
sinandcos = alt.Chart(x).transform_calculate(
    sin='sin(datum.x)',
    cos='cos(datum.x)'
)
```

Строим графики. В начале синуса, затем косинуса

```python
sin1 = sinandcos.mark_line().encode(
    x='x:Q',
    y='sin:Q',
    color=alt.value('red')
)

cos1 = sinandcos.mark_line().encode(
    x='x:Q',
    y='cos:Q',
    color=alt.value('blue')
)
sin1 | cos1
```

Объединяем графики

```python
sin1 + cos1
```

5.7. Используйте любимую визуализацию из предыдущего упражнения и настройте эстетику графика:


- настроить вид меток (`size`, `strokewidth` и т. д.).

- изменить оси и названия графика.

- изменить пределы `x` и `y`.

```python
# Создадим график со значениями синуса, которы имеет толщину 6, круглые концы, оранжевый цвет и включает всплывающую подсказку.
# также используем фильтрация, чтобы отобразить только тех точкек, где косинус больше или равен синусу
sin3 = sinandcos.mark_line(
    size=6,
    strokeCap='round',
    tooltip='cos'
).encode(
    x=alt.X('x:Q', title='Значение по оси X'),
    y=alt.Y('cos:Q', title='Значение Косинуса'),
    color=alt.value('orange')
).transform_filter(
    alt.datum.cos >= alt.datum.sin
)

# Создаим график со значениями косинуса, который имеет толщину 5, прозрачность 0.2, пунктирный, зеленый цвет и всплывающую подсказку.
# Nакже используем фильтрацию для отображения только точек, где косинус больше или равен синусу
cos3 = sinandcos.mark_line(
    size=5,
    opacity=0.2,
    strokeDash=[8, 2],
    tooltip='sin'
).encode(
    x=alt.X('x:Q', title='Значение по оси X'),
    y=alt.Y('sin:Q', title='Значение синуса'),
    color=alt.value('green')
).transform_filter(
    alt.datum.cos >= alt.datum.sin
)

# Объединим графики
sin3 + cos3
```
